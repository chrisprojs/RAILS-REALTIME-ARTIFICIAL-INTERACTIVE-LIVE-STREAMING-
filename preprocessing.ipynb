```python
import pandas as pd
import os

def list_files(directory):
  data = []
  for root, _, files in os.walk(directory):
    for file in files:
      directory = f"{file}"
      filename = os.path.splitext(file)[0]
      action = "run_video"
      data.append([directory, action, filename])
  df = pd.DataFrame(data, columns=["directory", "action", "label"])
  df.to_csv("label_AI_Live_Shopping.csv", index=False)

list_files("video_part")
```
